# Plotting the parameters for 15km detector
This document assumes the familiarity with plotting the sensitivity. See Run_example to learn more.

```python
 # Load the necessary packages
import matplotlib.pyplot as plt
import gwinc 
import numpy as np
%matplotlib inline
import ETLF.sqz_dual_fc as computeFCParamsLF # Needed to compute FC parameters for a given detector
import ETHF.squeezingFC as computeFCParamsHF
```

## LF 15km

```python
budgetLF = gwinc.load_budget('ETLF',freq=np.logspace(0,2,3000))  # freq defines the frequency range
budgetLF.ifo.Optics.SRM.Transmittance=0.152
budgetLF.ifo.Optics.SRM.Tunephase = -0.54
budgetLF.ifo.Optics.SRM.CavityLength = 168.6
budgetLF.ifo.Infrastructure.Length=15000

budgetLF.ifo.Optics.Curvature.ITM=10670
budgetLF.ifo.Optics.Curvature.ETM=10670
FCparams = computeFCParamsLF.computeFCParams(budgetLF.ifo)

budgetLF.ifo.Squeezer.FilterCavity[0] = FCparams[0]
budgetLF.ifo.Squeezer.FilterCavity[1] = FCparams[1]


tracesLFbase = budgetLF.run()


fig = gwinc.plot_budget(tracesLFbase)
plt.ylabel('Strain [$1 / \sqrt{\mathrm{Hz}}$]')
plt.ylim([5e-26, 1e-21])
```

```python
budgetLF = gwinc.load_budget('ETLF',freq=np.logspace(0,2,3000))  # freq defines the frequency range
FCparams = computeFCParamsLF.computeFCParams(budgetLF.ifo)
budgetLF.ifo.Squeezer.FilterCavity[0] = FCparams[0]
budgetLF.ifo.Squeezer.FilterCavity[1] = FCparams[1]
tracesLFbase = budgetLF.run()

budgetLF.ifo.Optics.SRM.Transmittance=0.152
budgetLF.ifo.Optics.SRM.Tunephase = -0.54
budgetLF.ifo.Optics.SRM.CavityLength = 168.6
budgetLF.ifo.Infrastructure.Length=15000

budgetLF.ifo.Optics.Curvature.ITM=10670
budgetLF.ifo.Optics.Curvature.ETM=10670
FCparams = computeFCParamsLF.computeFCParams(budgetLF.ifo)

budgetLF.ifo.Squeezer.FilterCavity[0] = FCparams[0]
budgetLF.ifo.Squeezer.FilterCavity[1] = FCparams[1]

tracesLF2L = budgetLF.run()

plt.loglog(np.logspace(0,4,3000), tracesLFbase.asd, label=f'LF 10km', color='green')
plt.loglog(np.logspace(0,4,3000), tracesLF2L.asd, label=f'LF 15km', color='brown')
plt.legend()
plt.grid(which='both', ls='--', alpha=0.8)
plt.ylabel('Strain [$1 / \sqrt{\mathrm{Hz}}$]')
plt.xlabel('Frequency [Hz]')
plt.ylim([4e-25, 5e-24])
plt.xlim([10,6000])
```

## HF 15km

```python
budgetHF = gwinc.load_budget('ETHF',freq=np.logspace(0,4,3000))  # freq defines the frequency range


budgetHF.ifo.Infrastructure.Length=15000
budgetHF.ifo.Optics.Curvature.ITM =7750
budgetHF.ifo.Optics.Curvature.ETM =7750
budgetHF.ifo.Optics.SRM.CavityLength = 170
budgetHF.ifo.Optics.SRM.Transmittance=0.033

FCparams = computeFCParamsHF.computeFCParams(budgetHF.ifo)
budgetHF.ifo.Squeezer.FilterCavity = FCparams
tracesHFbase = budgetHF.run()

fig = gwinc.plot_budget(tracesHFbase)
plt.ylabel('Strain [$1 / \sqrt{\mathrm{Hz}}$]')
plt.ylim([5e-26, 1e-21])
```

```python
budgetHF = gwinc.load_budget('ETHF',freq=np.logspace(0,4,3000))  # freq defines the frequency range
FCparams = computeFCParamsHF.computeFCParams(budgetHF.ifo)
budgetHF.ifo.Squeezer.FilterCavity = FCparams
tracesHFbase = budgetHF.run()


budgetHF.ifo.Infrastructure.Length=15000
budgetHF.ifo.Optics.Curvature.ITM =7750
budgetHF.ifo.Optics.Curvature.ETM =7750
budgetHF.ifo.Optics.SRM.CavityLength = 170
budgetHF.ifo.Optics.SRM.Transmittance=0.033

FCparams = computeFCParamsHF.computeFCParams(budgetHF.ifo)
budgetHF.ifo.Squeezer.FilterCavity = FCparams
tracesHF2L = budgetHF.run()

plt.loglog(np.logspace(0,4,3000), tracesHFbase.asd, label=f'HF 10km', color='green')
plt.loglog(np.logspace(0,4,3000), tracesHF2L.asd, label=f'HF 15km', color='brown')
plt.legend()
plt.grid(which='both', ls='--', alpha=0.8)
plt.ylabel('Strain [$1 / \sqrt{\mathrm{Hz}}$]')
plt.xlabel('Frequency [Hz]')
plt.ylim([1e-25, 5e-24])
plt.xlim([10,6000])
```

## Compare different designs

```python
etd = np.loadtxt("ET/et_d.txt")
plt.loglog(np.logspace(0,4,3000),etd[:,1], ls='--', label='ET-D', color='gray')


budgetHF = gwinc.load_budget('ETHF',freq=np.logspace(0,4,3000))  # freq defines the frequency range
## CoBA design based on the CoBA values
budgetHF.ifo.Infrastructure.Length = 10000
budgetHF.ifo.Optics.SRM.CavityLength = 100 # SRM transmittance
budgetHF.ifo.Materials.MassRadius = 0.31
budgetHF.ifo.Optics.Curvature.ITM = 5070
budgetHF.ifo.Optics.Curvature.ETM = 5070
budgetHF.ifo.Optics.SRM.Transmittance=0.1
budgetHF.ifo.Optics.BSLoss = 0.5e-3
budgetHF.ifo.Squeezer.InjectionLoss = 0.02
budgetHF.ifo.Squeezer.FilterCavity.L = 300
budgetHF.ifo.Squeezer.FilterCavity.Lrt = 55e-6
budgetHF.ifo.Squeezer.AmplitudedB = 15

FCparams = computeFCParamsHF.computeFCParams(budgetHF.ifo)
budgetHF.ifo.Squeezer.FilterCavity = FCparams
tracesHFbase = budgetHF.run()
plt.loglog(np.logspace(0,4,3000), tracesHFbase.asd, label=f'CoBA, 10km', ls = '-.', color='black')


budgetHF = gwinc.load_budget('ETHF',freq=np.logspace(0,4,3000))  # freq defines the frequency range
FCparams = computeFCParamsHF.computeFCParams(budgetHF.ifo)
budgetHF.ifo.Squeezer.FilterCavity = FCparams
tracesHFbase = budgetHF.run()
plt.loglog(np.logspace(0,4,3000), tracesHFbase.asd, label=f'Current design 10km', color='magenta')

budgetHF = gwinc.load_budget('ETHF',freq=np.logspace(0,4,3000))  # freq defines the frequency range

budgetHF.ifo.Infrastructure.Length=15000
budgetHF.ifo.Optics.Curvature.ITM =7750
budgetHF.ifo.Optics.Curvature.ETM =7750
budgetHF.ifo.Optics.SRM.Transmittance=0.033
budgetHF.ifo.Optics.SRM.CavityLength = 170
FCparams = computeFCParamsHF.computeFCParams(budgetHF.ifo)
budgetHF.ifo.Squeezer.FilterCavity = FCparams
tracesHFbase = budgetHF.run()
plt.loglog(np.logspace(0,4,3000), tracesHFbase.asd, label=f'Current design 15km', color='green')
plt.ylabel('Strain [$1 / \sqrt{\mathrm{Hz}}$]')
plt.ylim([1e-25, 5e-24])
plt.xlim([10,6000])
plt.legend()
```
