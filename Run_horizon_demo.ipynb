```python
import inspiral_range
import gwinc
import numpy as np
import matplotlib.pyplot as plt
```

```python
budgetET = gwinc.load_budget('ET')  # load the full noise budget
tracesET = budgetET.run()  # compute the noise budget
```

```python
Data=np.loadtxt("ETSensitivity_Scenarios.txt")  # load the ET-D sensitivity
fET=Data[:,12]  # get the ET-D frequency range
NET=Data[:,13]  # get the ET-D sensitivity
NETnew = tracesET.asd  # take amplitude spectral densitivy of the current budget
```

```python
M=np.linspace(0.5,500,600)  # define the mass range
```

```python
redETD=np.empty([600])  #  array to store computed red shift
redETnew=np.empty([600])
```

```python
### compute the red shift for each mass

for i in range(len(M)):
    rangeETD = inspiral_range.horizon_redshift(fET, NET**2/(1.5)**2, m1=M[i], m2=M[i])
    redETD[i]=rangeETD

    rangeETnew = inspiral_range.horizon_redshift(fET, NETnew**2/(1.5)**2,m1=M[i], m2=M[i])
    redETnew[i]=rangeETnew
```

```python
### Plot the comparsion of redshifts
fig = plt.figure(figsize=(7,4.3))
plt.loglog(2*M,redETD,lw=2)
plt.loglog(2*M,redETnew,lw=2)
plt.grid(which='both',ls='-',linewidth=0.5,alpha=0.5)
plt.legend(['ET-D','ET new'],ncol=1,loc='lower right')
plt.xlabel('Total Mass [$M_\odot$]')
plt.ylabel('Redshift')
plt.xlim(1,1000)
```
